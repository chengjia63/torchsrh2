```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
name_map = {
    "6e4fa842": "dinov2_lr0.004",
    "997de344": "dinov2_lr0.0004",
    "7679a1cb": "dinov2_lr0.0002",
    "c2aa6c21": "dinov2_lr0.0001",
    "06939699": "hidisc_lr0.0004",
    "baacd93b": "hidisc_lr0.0002",
    "22561ea5": "hidisc_lr0.0001",
    "b90b8a37": "hidisc_lr0.00005",
}
best_fm ={
    "tcga_patch": 0.701,
    "dbta_patch": 0.396,
    "umbtb_patch": 0.459,
    "tcga_slide": 0.828,
    "dbta_slide": 0.550,
    "umbtb_slide": 0.584
}
uni = {
    "tcga_patch": 0.695,
    "dbta_patch": 0.396,
    "umbtb_patch": 0.440,
    "tcga_slide": 0.805,
    "dbta_slide": 0.550,
    "umbtb_slide": 0.535
}
key = [["tcga_patch", "dbta_patch", "umbtb_patch"],
       ["tcga_slide", "dbta_slide", "umbtb_slide"]]
```

```python
results = {exp: pd.read_csv(f"data/{exp}.csv") for exp in name_map}
```

```python
def proc_ax(axi):
    axi.yaxis.set_ticks_position('none')
    axi.xaxis.set_ticks_position('none')
    axi.grid(zorder=0, color="lightgray")
    axi.set_xlim(0,350000)
    axi.set_xticks(np.arange(0, 350001, 50000))
    axi.set_xticklabels([f"{i//1000}k" for i in np.arange(0, 350001, 50000)])

    for w in ['top','bottom','left','right']:
        axi.spines[w].set_linewidth(1)
        axi.spines[w].set_zorder(0)
        axi.spines[w].set_color("lightgray")
    #ax.spines[w].set_visible(False)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(20,8))
for key_row, ax_row in zip(key, ax):
    for keyi, axi in zip(key_row, ax_row):
        axi.plot([0, 350000], [best_fm[keyi], best_fm[keyi]], label="best_fm", color="darkgray")

        for exp in name_map:
            axi.plot(np.hstack(([0], results[exp]["iter"])),
                     np.hstack(([uni[keyi]], results[exp][keyi])),
                     label=name_map[exp])

        axi.set_title(keyi)
        axi.legend()
        proc_ax(axi)

plt.rcParams['pdf.fonttype'] = 42
#plt.rcParams['font.family'] = "Helvetica Neue"

plt.tight_layout()
plt.savefig("dinov2_mca.pdf")
```
