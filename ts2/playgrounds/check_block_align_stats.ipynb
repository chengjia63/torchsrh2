```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
accepted = pd.read_csv("/nfs/turbo/umms-tocho/code/chengjia/torchsrh2/histreg/out/accepted.csv")
no_he = pd.read_csv("/nfs/turbo/umms-tocho/code/chengjia/torchsrh2/histreg/out/no_he.csv")
one_section = pd.read_csv("/nfs/turbo/umms-tocho/code/chengjia/torchsrh2/histreg/out/one_section.csv")
all_meta = pd.read_csv('/nfs/mm-isilon/brainscans/dropbox/Slide_Incoming/metadata/master_spreadsheet/matched_su_mxa.csv')
```

```python
counts = {
"Fitted affine / rigid\ntransform": (~(accepted["which"] == "block_align_manual")).sum(),
"Manual transform": (accepted["which"] == "block_align_manual").sum()
}
counts.update({"No H&E": len(no_he["reason"])})
counts.update({"One section": len(one_section["which"])})
```

```python
counts = pd.DataFrame.from_dict(counts, orient="index", columns=["count"])
total_blocks = len(all_meta[~ all_meta["Block"].str.endswith("x")].groupby(["UM Accession", "Block"]))

counts.loc["Verify pending", "count"] = 1272
counts.loc["In progress", "count"] = total_blocks - counts["count"].sum()
```

```python
counts = counts.iloc[[2,3,0,1,4, 5]]
counts["label"] = [f"{x} {int(i)}" for x,i in zip(counts.index, counts["count"])]
```

```python
plt.rcParams.update({'font.size': 16})
fig, (ax1) = plt.subplots(1,1,figsize=(8, 8))
wedges, texts = ax1.pie(counts["count"],
                        labels=counts["label"],
                        colors=["pink", "lightgrey", "lightblue", "yellow", "lightgreen", "orange"],
                        wedgeprops=dict(width=0.15), startangle=90)

ax1.set_title("Blocks")
plt.tight_layout()
matplotlib.rcParams['pdf.fonttype'] = 42 
#plt.savefig("data_quality.pdf")
```
