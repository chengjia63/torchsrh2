```python
import os
from glob import glob
from os.path import join as opj
from tensorflow.python.summary.summary_iterator import summary_iterator
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
```

```python
matplotlib.rcParams["pdf.fonttype"] = 42
matplotlib.rcParams["font.size"] = 16
matplotlib.rcParams["font.family"] = 'DejaVu Sans'
```

```python
def load_tb_one_exp(log_dir):
    print(log_dir)
    data = {}
    for event in summary_iterator(log_dir):
        for value in event.summary.value:
            tag = value.tag
            if tag not in data:
                data[tag] = []
            data[tag].append((event.step, value.simple_value))
    return data
def load_tb_data(log_dirs, tags, key):
    all_logs = {}
    for t, ld in zip(tags, log_dirs):
        tb_fnames = glob(opj(ld, "tb/version_0/events.out.tfevents.*"))
        assert len(tb_fnames) == 1
        all_logs[t] = np.array(load_tb_one_exp(tb_fnames[0])[key])
    return all_logs
```

```python
def plot_tb_data(data, xlim=None, ylim=None, num_x_major=10):
    # Create a plot for each tag
    fig, ax = plt.subplots(1,1, figsize=(12,9))
    for tag, values in data.items():
        ax.plot(values[:,0], values[:,1], label=tag, alpha=0.8)
    
    if xlim is None:
        xlim = [0, max(values[:,0].max()+1 for _, values in data.items())]
    
    ax.legend()
    ax.xaxis.set_tick_params(length=0)
    ax.yaxis.set_tick_params(length=0)
    ax.set_xlabel('Steps')
    ax.set_ylabel('Loss')

    ax.set_xlim(xlim)
    if xlim is not None:
        ax.set_ylim(ylim)
    ax.set_xticks(np.linspace(0,xlim[-1],num_x_major+1))
    ax.set_xticklabels([f"{int(i//1000)}K" for i in np.linspace(0,xlim[-1],num_x_major+1)])
    for s in ['right', 'top', "bottom", "left"]:
        ax.spines[s].set_color("lightgrey")
    ax.grid(c="lightgrey")
        
    return fig, ax
```

```python
log_root = "/nfs/turbo/umms-tocho-snr/exp/chengjia/ts2_scsrh/scsrh_supcon"
exps = [
    "84c6fb1d_Nov03-00-14-38_sd1000_simclr_lr0.001_tune4",
    "068d05c5_Nov01-03-17-31_sd1000_simclr_lr0.0001_tune0",
    "f69275a5_Nov01-17-34-55_sd1000_simclr_lr0.00001_tune2"
]
simclr_logs = load_tb_data(
    [opj(log_root, e) for e in exps],
    ["simclr_lr0.001", "simclr_lr0.0001", "simclr_lr0.00001"],
    "train/contrastive_step")
fig, ax = plot_tb_data(simclr_logs, ylim=[2, 7.5])
plt.tight_layout()
#fig.savefig("simclr.pdf")
```
