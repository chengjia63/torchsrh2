```python
import pandas as pd
import re
```

```python
data = pd.read_csv("mosaic.csv")
```

```python
data.groupby("ttype", group_keys=False).apply(lambda x: x.sample(frac=0.2))["ttype"].value_counts()
```

```python
data["nion"] = data["mosaic"].str.extract("(NIO_UM_[0-9]+)")
```

```python
valids = data[["ttype", "nion"]].groupby("ttype", group_keys=False).apply(lambda x: x.sample(frac=0.2, random_state=1000))
valid_nion = valids["nion"].tolist()
```

```python
data["ttype"].value_counts()
```

```python
data[data["nion"].isin(valid_nion)].drop("nion", axis=1).to_csv("mosaic_val.csv")
```

```python
data[~data["nion"].isin(valid_nion)].to_csv("mosaic_train.csv")
```
