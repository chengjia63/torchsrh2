```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color
import cv2

# Load the image
image_path = 'image.png'
image = io.imread(image_path)
image_gray = color.rgb2gray(image[...,:-1])

# Create a mask to neglect the white background
mask = image[...,-1] > 0  # Threshold to separate tissue from background
mask = cv2.morphologyEx(mask.astype(np.uint8), cv2.MORPH_OPEN, np.ones((8,8), np.uint8)).astype(bool)
# Generate random points within the mask
np.random.seed(0)
points = []
for _ in range(64):
    while True:
        x = np.random.randint(0, image.shape[0])
        y = np.random.randint(0, image.shape[1])
        if mask[x, y]:
            points.append((x, y))
            break

# Generate a Gaussian distribution around these points
x = np.linspace(0, image.shape[1], image.shape[1])
y = np.linspace(0, image.shape[0], image.shape[0])
X, Y = np.meshgrid(x, y)

Z = np.zeros_like(X, dtype=float)
sigma = 80  # Standard deviation for the Gaussian

for (x0, y0) in points:
    Z += np.exp(-((X - y0)**2 + (Y - x0)**2) / (2 * sigma**2))

Z[~mask] = -1
```

```python
# Plot the contour plot
plt.figure(figsize=(10, 10))
plt.contour(X, Y, Z, levels=20, zorder=3)
#plt.imshow(Z)
plt.imshow(image, zorder=0, alpha=.5)
plt.axis('off')
```
