```python
import fitz
import numpy as np
import qrcode
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
import einops
```

```python
def generate_qr_with_text(data, qr_size=100):
    # Generate QR code
    qr = qrcode.QRCode(border=2)  # Small border for fitting
    qr.add_data(data+"_mlins")
    qr.make(fit=True)
    
    # Create QR image
    qr_img = qr.make_image(fill="black", back_color="white").convert("RGB")
    
    # Resize QR code to desired size
    qr_img = qr_img.resize((qr_size, qr_size))
    
    # Add text below QR code
    img_with_text = Image.new("RGB", (qr_size, qr_size + 20), "white")  # Extra space for text
    img_with_text.paste(qr_img, (0, 0))
    
    # Add text
    draw = ImageDraw.Draw(img_with_text)
    #font = ImageFont.load_default()  # Use default font or provide your own TTF file
    font = ImageFont.truetype("assets/RobotoMono-VariableFont_wght.ttf", size=15)

    text_bbox = draw.textbbox((0, 0), data, font=font)
    text_width = text_bbox[2] - text_bbox[0]  # width = right - left
    text_x = (qr_size - text_width) // 2
    
    draw.text((text_x, qr_size-5), data, fill="black", font=font)
    
    return img_with_text
```

```python
def get_linear_im_one_page(a_nums,n_code_per_sticker=3, n_sticker_cols=3): #a code
    all_im = []
    a_nums = einops.rearrange(a_nums,"(sc c sr) -> (sc sr) c", c=n_code_per_sticker, sc=n_sticker_cols)
    for sticker in a_nums:
        mx = [f"A{ai}" for ai in sticker]
        all_im.append(np.hstack([np.array(generate_qr_with_text(mxi)) for mxi in mx]))
    return all_im
```
