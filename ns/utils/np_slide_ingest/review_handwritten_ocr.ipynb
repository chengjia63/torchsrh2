```python
import openslide
import matplotlib.pyplot as plt
import numpy as np
import glob
import json
import pandas as pd
from tqdm import tqdm
```

```python
def read_cat_csvs(fn_pattern:str) -> pd.DataFrame:
    meta_fns = glob.glob(fn_pattern)
    return pd.concat([pd.read_csv(mfn) for mfn in meta_fns]).reset_index(drop=True)
```

```python
ocr_fns = glob.glob("scanning_data/2024*v2.csv")
ocr_metas = pd.concat([pd.read_csv(mfn) for mfn in ocr_fns]).reset_index(drop=True)
```
